krx api를 이용한 주식 정보 조회

```python
import requests
import json

url = ('http://data-dbg.krx.co.kr/svc/apis/sto/stk_bydd_trd'  # URL
       + '?basDd='  # request inBlock
       + '20230707')  # yyyymmdd
headers = {
    'AUTH_KEY': '키번호'  # sample auth key
}
res = requests.get(url=url, headers=headers)
print('200 is ok :', res.status_code)
print(res.text)
```

res.text는 이미 문자열 형식으로 되어 있어 json.loads()을 할 필요가 없다.

```python
res = json.loads(res.text)
print(type(res))
```

```python
print(res['OutBlock_1'][3])
```

res['OutBlock_1'] 안에 딕셔너리의 키를 변경하는게 목표이다. old_key는 'BAS_DD' : 'new_key'는 기준일자 이다

```python
translated_keys = {
    'BAS_DD': '기준일자',
    'ISU_CD': '종목코드',
    'ISU_NM': '종목명',
    'MKT_NM': '시장명',
    'SECT_TP_NM': '소속부',
    'TDD_CLSPRC': '종가',
    'CMPPREVDD_PRC': '전일대비',
    'FLUC_RT': '등락률',
    'TDD_OPNPRC': '시가',
    'TDD_HGPRC': '고가',
    'TDD_LWPRC': '저가',
    'ACC_TRDVOL': '거래량',
    'ACC_TRDVAL': '거래대금',
    'MKTCAP': '시가총액',
    'LIST_SHRS': '상장주식수'
}

# res['OutBlock_1'] 안에 있는 딕셔너리들에 대해 작업
for item in res['OutBlock_1']:
    # 각 딕셔너리의 키를 변경
    for old_key, new_key in translated_keys.items():
        # 만약 old_key가 딕셔너리 안에 존재한다면
        if old_key in item:
            # 해당 딕셔너리에서 키를 변경
            item[new_key] = item.pop(old_key)

# 출력하여 변경된 결과 확인
print(res['OutBlock_1'][1])
```

예외처리로 data가 {} 일때 csv추가 금지 해야함

```python
import csv

# 사용자로부터 종목명 입력 받기
st_name = input("종목명을 입력하세요")

# 검색된 데이터를 저장할 변수
found_data = None

# 데이터를 CSV 파일에 추가하기 전에 검색된 데이터가 있는지 확인
for test in res['OutBlock_1']:
    if test.get('종목명') == st_name:
        found_data = test
        break

# 검색된 데이터가 있을 경우에만 처리
if found_data is not None:
    # CSV 파일 열기 (쓰기 모드)
    with open('data.csv', 'a', newline='') as csvfile:
        # CSV 파일에 데이터를 작성할 writer 객체 생성
        writer = csv.DictWriter(csvfile, fieldnames=found_data.keys())

        # CSV 파일에 헤더를 추가할지 확인
        if csvfile.tell() == 0:
            writer.writeheader()

        # 데이터를 CSV 파일에 추가
        writer.writerow(found_data)

        # 검색된 데이터 출력
        for key, value in found_data.items():
            print(f"{key}: {value}")
else:
    print("해당 종목을 찾을 수 없습니다.")
```
