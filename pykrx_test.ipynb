OHLCV를 정리하면

- **O (Open):** 주식이 특정 기간(예: 하루)의 거래를 시작할 때의 가격입니다.
- **H (High):** 특정 기간 동안의 최고 가격입니다.
- **L (Low):** 특정 기간 동안의 최저 가격입니다.
- **C (Close):** 특정 기간(예: 하루)의 거래가 마감할 때의 가격입니다.
- **V (Volume):** 특정 기간 동안의 거래된 주식 수입니다.

IV (Intrinsic Value, 내재 가치): 회사의 주식이 현재 시장 가격보다 가치가 높거나 낮은지를 나타내는 지표입니다. IV는 다양한 재무 지표와 적정 주가를 고려하여 산출될 수 있으며, 주식 시장에서 주가와의 차이를 분석하는 데 사용됩니다.

BPS (Book Value Per Share, 주당 자산가치): 회사의 순자산을 발행된 주식 수로 나눈 값으로, 주식이 현재의 시장 가격보다 실제로 가치 있는지를 판단하는 데 사용됩니다. BPS는 보통 주식의 내재 가치를 평가하는 데에 사용됩니다.

PER (Price-to-Earnings Ratio, 주가수익비율): 주식 시장에서 주가를 기반으로 한 회사의 이익을 나타내는 지표로, 주가를 기반으로 하는 주식 가격을 기업의 순이익으로 나눈 것입니다. PER은 투자자가 주식을 구입할 때 기업의 수익을 평가하는 데 사용됩니다. 일반적으로 낮은 PER은 저평가된 주식을 나타낼 수 있으며, 높은 PER은 고평가된 주식을 나타낼 수 있습니다.

EPS (Earnings Per Share, 주당순이익): 기업의 순이익을 발행된 주식 수로 나눈 값으로, 각 주식에 대한 순이익을 나타냅니다. EPS는 기업의 수익성을 측정하는 데 사용되며, 높은 EPS는 투자자에게 주식을 보다 유망한 투자로 인식시킬 수 있습니다.

코스피 인덱스를 조회합니다.

```python
tickers = stock.get_index_ticker_list()
print(tickers)
```

```python
import matplotlib.pyplot as plt
from pykrx import stock

# 티커로부터 종목명 얻기
def get_stock_name_by_ticker(ticker):
    stock_names = stock.get_market_ticker_name(ticker)
    return stock_names if stock_names else None

# 티커로 주가 데이터 가져오기
def get_stock_price_by_ticker(ticker, start_date, end_date):
    df = stock.get_market_ohlcv_by_date(start_date, end_date, ticker)
    return df

# 사용자에게 티커 입력 받기
ticker = input("검색할 종목의 티커를 입력하세요: ")



# 종목명 가져오기
stock_name = get_stock_name_by_ticker(ticker)

# 종목명이 없을 경우 종료
if stock_name is None:
    print("해당하는 종목이 없습니다.")
    exit()

# 주가 데이터 가져오기
df = get_stock_price_by_ticker(ticker, "20230101", "20240306")

# 그래프 그리기
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['종가'], label='종가', color='blue')
plt.title('{} 종목 주가 변화'.format(stock_name))
plt.xlabel('날짜')
plt.ylabel('주가 (원)')
plt.legend()
plt.grid(True)
plt.show()

print(stock_name)
```
